```python
from jupyter_dash import JupyterDash
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic


```

```python
df = pd.read_csv("India-Tourism-Statistics-2021-Table-2.1.2_0.csv")
df.head(15)
```

```python
# Initialize the app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    
    html.H1("Data Visualization Dashboard for India Tourism Statistics 2018-2020"),
    
    # Histogram
    dcc.Graph(
        id='histogram',
        figure={
            'data': [{
                'x': df['Growth 2020/19 (%)'],
                'type': 'histogram'
            }],
            'layout': {
                'title': 'Histogram',
                 'xaxis' : {'title' : 'Growth 2020/19 (%)'}
               
            }
            }
        
    ),
    
    # Box plot
    dcc.Graph(
        id='boxplot',
        figure={
            'data': [{
                'x': df['Months'],
                'y': df['2018'],
                'type': 'box'
            }],
            'layout': {
                'title': 'Box plot',
                'xaxis' : {'title' : 'Months'},
                'yaxis' : {'title' : '2018'}
                
            }
        }
    ),
    # Mosaic plot
    dcc.Graph(
        id='mosaicplot',
        figure={
            'data': [{
                'y': df['2020'],
                'type': 'mosaicplot'
            }],
            'layout': {
                'title': 'Mosaic plot',
                'yaxis' : {'title' : '2020'}
            }
        }
    ),
    
    # Scatter plot
    dcc.Graph(
        id='scatterplot',
        figure={
            'data': [{
                'x': df['Growth 2019/18 (%)'],
                'y': df['Growth 2020/19 (%)'],
                'mode': 'Growth',
                'marker': {
                    'size': 15,
                    'color': 'blue',
                    'opacity': 0.7
                }
            }],
            'layout': {
                'title': 'Scatter plot',
                'xaxis' : {'title' : 'Growth 2019/18 (%)'},
                'yaxis' : {'title' : 'Growth 2020/19 (%)'}
                
            }
        }
    ),
    
    # Bar chart
    dcc.Graph(
        id='barchart',
        figure={
            'data': [{
                'x': df['Months'],
                'y': df['2020'],
                'type': 'bar'
            }],
            'layout': {
                'title': 'Bar chart',
                'xaxis' : {'title' : 'Months'},
                'yaxis' : {'title' : '2020'}
                
                
            }
        }
    )
])


```

```python
# Run the app
if __name__ == '__main__':
    app.run_server(port=3050)
```
